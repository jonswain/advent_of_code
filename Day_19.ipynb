```python
day19 = (open('inputs/day19.txt').read().split('\n'))
```

'Blueprint 1: Each ore robot costs 4 ore. Each clay robot costs 4 ore. Each obsidian robot costs 4 ore and 17 clay. Each geode robot costs 4 ore and 16 obsidian.',

```python
blueprints = [
    {'blueprint': 1,
 'or_robot': 4,
 'c_robot': 2,
 'ob_robot_ore': 3,
 'ob_robot_c': 14,
 'g_robot_ore': 2,
 'g_robot_obs': 7},
    {'blueprint': 2,
 'or_robot': 2,
 'c_robot': 3,
 'ob_robot_ore':3,
 'ob_robot_c': 8,
 'g_robot_ore': 3,
 'g_robot_obs': 12}
]
```

```python
blueprints = []


for line in day19:
    blueprint = int(line.split()[1].strip(':'))
    or_robot = int(line.split()[6])
    c_robot = int(line.split()[12])
    ob_robot_ore = int(line.split()[18])
    ob_robot_c = int(line.split()[21])
    g_robot_ore = int(line.split()[27])
    g_robot_obs = int(line.split()[30])
    blueprints.append({
        'blueprint' : blueprint,
        'or_robot' : or_robot,
        'c_robot' : c_robot,
        'ob_robot_ore' : ob_robot_ore,
        'ob_robot_c' : ob_robot_c,
        'g_robot_ore' : g_robot_ore,
        'g_robot_obs' : g_robot_obs
    })
    
blueprints[0]
```

```python
qual_levels = []



for blueprint in blueprints:
    time = 24
    ore_robots = 1
    clay_robots = 0
    obs_robots = 0
    geo_robots = 0
    
    ore = 0
    clay = 0
    obs = 0
    geo = 0
    
    while time > 0:
        
        print(f"Starting minute {time}, you have {ore_robots} ore robots, {clay_robots} clay robots {obs_robots} obsidian robots, and {geo_robots} geode robots")
        print(f"You have {ore} ore, {clay} clay, {obs} obsidian, and {geo} geodes")
        ore += ore_robots
        clay += clay_robots
        obs += obs_robots
        geo += geo_robots
        print(f"After mining, you have {ore} ore, {clay} clay, {obs} obsidian, and {geo} geodes")
        
        if (ore >= blueprint['g_robot_ore']) & (obs >= blueprint['g_robot_obs']):
            ore -= blueprint['g_robot_ore']
            obs -= blueprint['g_robot_obs']
            geo_robots += 1
            print(f"Buying one geode robot for {blueprint['g_robot_ore']} ore and {blueprint['g_robot_obs']} obsidian")        
            
        elif (ore >= blueprint['ob_robot_ore']) & (clay >= blueprint['ob_robot_c']):
            ore -= blueprint['ob_robot_ore']
            clay -= blueprint['ob_robot_c']
            obs_robots += 1
            print(f"Buying one obsidian robot for {blueprint['ob_robot_ore']} ore and {blueprint['ob_robot_c']} clay")
            
        elif ore >= blueprint['c_robot']:
            ore -= blueprint['c_robot']
            clay_robots += 1
            print(f"Buying one clay robot for {blueprint['c_robot']} ore")            
            
        elif ore >= blueprint['or_robot']:
            ore -= blueprint['or_robot']
            ore_robots += 1
            print(f"Buying one ore robot for {blueprint['or_robot']} ore")                
        
        time -= 1
        print()
        
    qual_level = blueprint['blueprint']*geo
    qual_levels.append(qual_level)

print(sum(qual_levels))
```
