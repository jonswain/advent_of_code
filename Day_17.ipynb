```python
day17 = list(open('inputs/day17.txt').read())
instruct_len = len(day17)
day17 = day17*1000
```

```python
import numpy as np

def new_block(block_num, y):
    block_num += 1
    if block_num%5 == 0:
        return [(2,y+4), (3,y+4), (2,y+3), (3,y+3)], block_num
    elif block_num%5 == 1:
        return [(2,y+3), (3,y+3), (4,y+3), (5,y+3)], block_num
    elif block_num%5 == 2:
        return [(2,y+4), (3,y+4), (3,y+3), (3,y+5), (4,y+4)], block_num
    elif block_num%5 == 3:
        return([(2,y+3), (3,y+3), (4,y+3), (4,y+4), (4,y+5)]), block_num
    elif block_num%5 == 4:
        return [(2,y+3), (2,y+4), (2,y+5), (2,y+6)], block_num
    
  
def move_left(block):
    if block[0][0] == 0:
        return block
    else: 
        for (x,y) in block:
            if cave[(x-1,y)] == 1:
                return block        
        block = [(x-1,y) for (x,y) in block]
        return block
    

def move_right(block):
    if block[-1][0] == 6:
        return block
    else: 
        for (x,y) in block:
            if cave[(x+1,y)] == 1:
                return block
        block = [(x+1,y) for (x,y) in block]
        return block
   

cave = np.full((7, 10000),0)
cave[:,0] = 1
block_num = 0
y=1
falling = False
block = [(0,0)]
answer = 0

for instruction in day17:
    if falling == False:
        
        for square in block:
            cave[square] = 1

        for index, row in enumerate(cave.T):
            if row.sum() == 0:
                y = index
                break        

        if block_num == 2022:
            answer = block
            break
        block, block_num = new_block(block_num, y)
        falling = True
          
    if instruction == "<":
        block = move_left(block)
        
    else: 
        block = move_right(block)

    for (x,y) in block:
        if cave[(x,y-1)] == 1:
            falling = False

        
    if falling == True:       
        block = [(x,y-1) for (x,y) in block]

print(f"Part 1 answer: {max([y for x,y in answer])}")
```
