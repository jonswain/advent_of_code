```python
import numpy as np
import collections
import itertools
from math import prod

day23 = (open('inputs/day23.txt').read())
day23 = day23.split('\n')
day23 = [list(line) for line in day23]
day23 = [[0 if char == '.' else 1 for char in line] for line in day23]
day23 = np.array(day23)
```

```python
for i in range(1000):
    if (day23[0].sum() > 0) or (day23[-1].sum() > 0) or (day23[:,0].sum() > 0) or (day23[:,-1].sum() > 0):
        day23 = np.pad(day23, 1)
    new_positions = {}
    non_move = 0
    
    for y, x in itertools.product(range((day23.shape[0])-2),range((day23.shape[1])-2)):
        y += 1
        x += 1
        if day23[y,x] == 1:
        
            N = day23[y-1,x]
            NE = day23[y-1,x+1]
            NW = day23[y-1,x-1]
            S = day23[y+1,x]
            SE = day23[y+1,x+1]
            SW = day23[y+1,x-1]
            W = day23[y,x-1]
            E = day23[y,x+1]
            
            if i%4 == 0:
                if sum([N,S,E,W,NE,NW,SE,SW]) == 0:
                    new_position = y, x
                    non_move += 1
                elif sum([N, NE, NW]) == 0:
                    new_position = y-1, x
                elif sum([S, SE, SW]) == 0:
                    new_position = y+1, x
                elif sum([W, NW, SW]) == 0:
                    new_position = y, x-1
                elif sum([E, NE, SE]) == 0:
                    new_position = y, x+1
                else:
                    new_position = y, x
                    non_move += 1
                    
            elif i%4 == 1:
                if sum([N,S,E,W,NE,NW,SE,SW]) == 0:
                    new_position = y, x
                    non_move += 1
                elif sum([S, SE, SW]) == 0:
                    new_position = y+1, x
                elif sum([W, NW, SW]) == 0:
                    new_position = y, x-1
                elif sum([E, NE, SE]) == 0:
                    new_position = y, x+1
                elif sum([N, NE, NW]) == 0:
                    new_position = y-1, x                    
                else:
                    new_position = y, x
                    non_move += 1
                    
            elif i%4 == 2:
                if sum([N,S,E,W,NE,NW,SE,SW]) == 0:
                    new_position = y, x
                    non_move += 1
                elif sum([W, NW, SW]) == 0:
                    new_position = y, x-1
                elif sum([E, NE, SE]) == 0:
                    new_position = y, x+1
                elif sum([N, NE, NW]) == 0:
                    new_position = y-1, x
                elif sum([S, SE, SW]) == 0:
                    new_position = y+1, x
                else:
                    new_position = y, x
                    non_move += 1
                    
            else:
                if sum([N,S,E,W,NE,NW,SE,SW]) == 0:
                    new_position = y, x
                    non_move += 1
                elif sum([E, NE, SE]) == 0:
                    new_position = y, x+1
                elif sum([N, NE, NW]) == 0:
                    new_position = y-1, x
                elif sum([S, SE, SW]) == 0:
                    new_position = y+1, x
                elif sum([W, NW, SW]) == 0:
                    new_position = y, x-1
                else:
                    new_position = y, x
                    non_move += 1

            new_positions[y,x] = new_position
        
    if non_move == day23.sum():
        print(f'Part 2 answer: {i+1}')
        break
        
    duplicates = [item for item, count in collections.Counter(new_positions.values()).items() if count > 1]
    for key, value in new_positions.items():
        if value in duplicates:
            new_positions[key] = key

    day23 = np.full(day23.shape, 0)
    for value in new_positions.values():
        day23[value] = 1
    
    if i == 9:
        rows = []
        cols = []

        for index, row in enumerate(day23):
            if row.sum() > 0:
                rows.append(index)

        for index, col in enumerate(day23.T):
            if col.sum() > 0:
                cols.append(index)

        print(f"Part 1 answer: {prod(day23[min(rows):max(rows)+1,min(cols):max(cols)+1].shape) - day23.sum()}")
        
```
